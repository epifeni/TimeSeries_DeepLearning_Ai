## Set the index of a dataframe

```python
# set the year to be a datetime index using statsmodels built-in functionality
# this is not often used now that pandas has robust datetime functionality
index = pd.Index(sm.tsa.datetools.dates_from_range('1959Q1', '2009Q3'))
```

```python
df.index = index
```

```python
df.head()
```

```python
df['realgdp'].plot()
plt.show()
```

### Seperate time series into trend and cyclical components

```python
results = sm.tsa.filters.hpfilter(df['realgdp'])
```

```python
type(results)
```

```python
results[0]  # estimated cyclical component
```

```python
results[1]  # estimated trend component
```

```python
# tuplie unpacking to get trend and cyclical components
gdp_cycle, gdp_trend = sm.tsa.filters.hpfilter(df['realgdp'])
```

```python
# add in the trend values to the dataframe
df['trend'] = gdp_trend
```

```python
# plot the original series and the trend
df[['realgdp', 'trend']].plot()
plt.show()
```

```python
df[['realgdp', 'trend']]["2000-03-31":].plot()
plt.show()
```
