### Random numbers

```python
np.random.rand(1) # random number between 0 and 1 uniform distribution. Meaning all numbers are equally likely ofbeing chosen
```

```python
np.random.rand(4, 5)
```

```python
# generate rnumbers from a standard normal distribution (mean 0, variance 1)
np.random.randn(5) # 5 random numbers randomly chosen from standard normal distribution
```

```python
np.random.randn(5, 4)
```

```python
# random integer from low (inclusive) to high (exclusive)
np.random.randint(1, 100)
```

```python
np.random.randint(1, 100, 10)
```
